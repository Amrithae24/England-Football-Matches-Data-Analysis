## 11.plotting Different features against one another(Boxplot,Lineplot,Countplot and scatter)

# 1.Boxplot

A boxplot is a graph that gives a visual indication of how a data set's 25th percentile, 50th percentile, 75th percentile, minimum, maximum and outlier values are spread out and compare to each other.

```python
sns.boxplot(x=df['Home_corner'], y=df['Away_corner'], data=df)
plt.show()
```

In the boxplot the higher corner is seen in Away corner with 14 than home corner

# 2.Lineplot

Line plots are constructed using several elements: a title to show what information is found on the plot, an x-axis where the values are listed, a caption to explain what the values on the x-axis are, and Xs above the values showing the number of times each one occurs.

```python
grouped = df.groupby(df['Home_corner'])
a=grouped['Total_corners'].agg(np.mean)
sns.lineplot(a)
plt.show()
```

From the above gragh we can observe that the high home corner is above 20.

## 3.Countplot

Countplot is basically a count of frequency plot in form of a bar graph. It plots the count of each category in a separate bar. When we use the pandas' value counts function on any column, It is the same visual form of the value counts function.

```python
sns.countplot(x=df['Away_corner'], data=df)
plt.show()
```

 The highest match of count is 3500 and the lowest is 0.5

# 4.scatter plot

The scatter plot is a mainstay of statistical visualization.
It depicts the joint distribution of two variables using
a cloud of points, where each point represents an
observation in the dataset. This depiction allows the eye
to infer a substantial amount of information about whether
there is any meaningful relationship between them. It is
plotted using the scatterplot() method.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Home_goal'], df['Total_corners'])
ax.set_xlabel('Home_goal')
ax.set_ylabel('Total_corners')
plt.show()
```

The above gragh shows that most of the result are lies in between 0-6.

# 5.Histogram

A histogram is a type of chart that shows the frequency distribution of data points across a continuous range of numerical values. The values are grouped into bin or buckets that are arranged in consecutive order along the horizontal x-axis at the bottom of the chart.

```python
df.Home_corner.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Home_goal")
plt.ylabel('Away_goal')
plt.xlabel('Home_corner');
```

The above gragh shows that the home corner is high when it is above 3000.
